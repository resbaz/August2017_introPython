Packages tend to contain functions that other programmers thought would be useful to use at some point. Things like, say, counting how many times particular words occur?

```python
import nltk
import collections
from nltk.tokenize.casual import casual_tokenize
```

```python
with open('Zen_of_Python.txt', 'r') as f:
    text = f.read()
```

```python
text = text.lower()
text
```

```python
text_tok = casual_tokenize(text)
```

```python
text_tok
```

```python
from collections import Counter

#Counter lets you quickly count how many times a word occurs
```

```python
Counter('hello trainees, how are you')
```

```python
Counter(["What",'about','a', 'list'])
```

```python
text_tok_counted = Counter(text_tok)
```

```python
text_tok_counted
```

```python
text_tok_counted.most_common(10)
```

Another way to do this, while removing any punctuation characters, is by removing the items from the dictionary, or creating a new one without the offending characters

```python
import string
```

```python
string.punctuation
```

```python
Counter('hello trainees, how are you')
```

```python
temp = Counter('hello trainees, how are you')
tlist = Counter()
type(temp)

for k,v in temp.items():
    if (k not in string.punctuation) and (k != " "):
        tlist[k] = v
        
print(tlist)
```
